```python
x=np.random.randn(100)
y=np.random.randn(100)
plt.title('산점도')
plt.scatter(x,y)
plt.show()
```

```python
# np.random.rand(개수)  -> 0~1까지 값 랜덤으로
# np.random.randn(개수) -> 평균 0, 표준편차 1의 표준정규분포 난수 값 랜덤으로
# np.random.randint(시작,끝, 갯수)
```

jitter graph = 중복된 값을 겹쳐서 부피가 커보이게 하는 그래프

```python
colors=np.random.rand(100)
area = np.pi * (15* np.random.rand(100))**2
plt.scatter(x,y, s=area, c=colors, alpha=0.5)
plt.title('버블티 먹고싶다')
plt.show()
```

### histogram

```python
s= np.random.randn(1000000)
plt.hist(s, bins=60) #bins는 기본 10개로 쪼개서 나오는 데이터를 20으로 늘여줌
plt.title('bins 20개인 경우')
plt.show()
```

### box plot

```python
# setoda 품종
iris = sns.load_dataset('iris')
setosa = iris[iris.species=='setosa']
```

```python
plt.figure(figsize=(8,10))
plt.boxplot([setosa.sepal_length, setosa.sepal_width, setosa.petal_length, setosa.petal_width],
labels=iris.columns[:-1])
plt.title('setosa 품종의 각 피쳐별 박스 플롯', fontsize=10)
plt.show()
```

```python
ratio = [34,32,16,18]
labels = ['사과','배','복숭아','바나나']

plt.pie(ratio, labels=labels, autopct='%.1f%%')
plt.show()
```
