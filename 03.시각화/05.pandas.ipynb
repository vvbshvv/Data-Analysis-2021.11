# 판다스 시각화

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus']=False
plt.rcParams['font.family']= 'Malgun Gothic'
```

```python
np.random.seed(2021)
df=pd.DataFrame(np.random.randn(100,3),index=pd.date_range('1/1/2021',periods=100),columns=list('ABC')).cumsum()

df.plot()
plt.title('판다스 프롯메서드 사용예')
plt.xlabel('시간')
plt.ylabel('데이터')
plt.show()
```

### bar chart

```python
import seaborn as sns
iris = sns.load_dataset('iris')

plt.figure(figsize=(12,6))
iris.sepal_length[:50].plot(kind='bar', rot=0)  ## sepal_length 시리즈 데이터를 바 그래프로 나타냄, rot=0 는 가로 인덱스 값을 세로로 나타냄
plt.title('Setosa 품종의 sepal length')
plt.show()
```

```python
df2 = iris.groupby('species').mean()
df2.columns.name='Feature'
df2
```

```python
df2.plot.bar(rot=0)
plt.title('Iris 각 품종의 feature별 평균')
plt.ylim(0,9) # 범례가 바 그래프를 가려 y축 범위를 더 넓게 지정
plt.show()

#plt.xlim()
```

### pie chart

```python
titanic = sns.load_dataset('titanic')
s = titanic['class'].value_counts()
s
```

```python
s.plot.pie(autopct='%.2f%%')
plt.title('선실별 승객수')
plt.show()
```

```python
titanic = sns.load_dataset('titanic')
s = titanic.groupby('class').mean()
s
```

```python
s.boxplot(rot=0)
plt.show()
```

### box plot

```python
iris[iris.species == 'setosa'].plot.box()
plt.title('setosa품종의 각 feature별 box plot')

plt.show()
```

```python
iris.boxplot(by='species', figsize=(10,8))
plt.show()
```
