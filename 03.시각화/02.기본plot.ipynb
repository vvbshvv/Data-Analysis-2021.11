```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus']=False
plt.rcParams['font.family']= 'Malgun Gothic'
```

```python
plt.plot([0,1,2,4],[2,4,6,8])
plt.title('라인 플롯')
plt.grid()
plt.show()
```

```python
plt.plot([0,1,2,4],[2,4,6,8], 'ro-.')  #'color, maker, linestyle'
plt.title('섹시한 빨간 동그라미를 가진 그래프')
plt.grid()
plt.show()
```

```python
plt.plot([0,5,7,9],[5,2,4,8], 'k*--')
plt.plot([1,2,5,8], 'gv:')
plt.title('여러가지 그래프, green')
plt.grid()
plt.show()
```

```python
plt.plot([0,5,7,9],[5,2,4,8], 'k*--', linewidth=3, markersize=10)
plt.plot([1,2,5,8], 'gv:', lw=5, ms=20) # lw 선의 폭, ms 마커의 크기
plt.title('오얄스')
plt.grid()
plt.show()
```

- 그림범위

```python
plt.plot([0,1,2,4],[2,4,6,8], 'ro-.')  #'color, maker, linestyle'
plt.title('섹시한 빨간 동그라미를 가진 그래프')
plt.xlim(-1,5)  #x축의 범위
plt.ylim(-1,10) #y축의 범위
plt.grid()
plt.show()
```

- tick 설정

```python
x = np.linspace(-np.pi, np.pi, 500)
y = np.cos(x)
plt.title('tick 설정')
plt.plot(x,y,'g*--', lw=5.5)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid()
plt.show()
```

- 범례

```python
X = np.linspace(-np.pi, np.pi, 500)
C,S = np.cos(X), np.sin(X)
plt.title('Legend를 표시한플롯')
plt.plot(X,C,label='cos',lw=5.5) #label을 이용해 범례 표시
plt.plot(X,S,label='sin',lw=5.5)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi]) # x축에 나타낼 축 값표시
plt.yticks([-1,0,1])
plt.grid()
plt.legend()
plt.show()
```

- x축 y축 label

```python
X = np.linspace(-np.pi, np.pi, 500)
C,S = np.cos(X), np.sin(X)
plt.title('x축 y축 이름 등록')
plt.plot(X,C,label='cos',lw=5.5)
plt.plot(X,S,label='sin',lw=5.5)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid(),plt.legend()
plt.xlabel('time'),plt.ylabel('Amplitude') # 축 제목 표시
plt.show()
```

- 그림 크기

```python
plt.figure(figsize=(10,6)) # 그림의 크기
plt.title('그림크기 지정')
plt.plot(X,C,label='cos',lw=5.5)
plt.plot(X,S,label='sin',lw=5.5)
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi])
plt.yticks([-1,0,1])
plt.grid(),plt.legend()
plt.xlabel('time'),plt.ylabel('Amplitude') # 축 제목 표시
plt.savefig('삼각함수.png')
plt.show()
```
