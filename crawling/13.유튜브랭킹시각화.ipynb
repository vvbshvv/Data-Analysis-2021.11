```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import matplotlib as mpl
mpl.rcParams['axes.unicode_minus']=False
plt.rcParams['font.family']= 'Malgun Gothic'
```

```python
df=pd.read_csv('유튜브 랭킹.csv')
df.head()
```

```python
df['구독자수']=df.구독자수.astype(int)
df['총조회수']=df.총조회수.astype(np.int64
)
df['비디오수']=df.비디오수.astype(int)
df.head()
```

```python
plt.figure(figsize=(12,8))
plt.barh(df.채널명.head(20), df.구독자수.head(20))
plt.title('구독자수 top 20채널')
plt.grid()
plt.show()
```

```python
df2 = df[['채널명','구독자수']].head(20)
df2.set_index('채널명', inplace=True)
df2.plot(kind='barh', grid=True, figsize=(12,8))
plt.title('구독자수 top 20채널')
plt.show()
```

```python
df2.sort_values(by='구독자수').plot(kind='barh', grid=True, figsize=(12,8))
plt.title('구독자수 top 20채널')
plt.show()
```

```python
df5=df.캐테고리.value_counts().to_frame().head(20)

plt.figure(figsize=(12,8))
plt.pie('캐테고리', labels=df5.index, data=df5, autopct='%.1f%%')
plt.title('카테고리별 채널 수')
plt.show()

```

- 카테고리별 구독자 수

```python
#df.캐테고리.value_counts('구독자수').to_frame().head(20)
df6=df[['캐테고리','구독자수']]
df7=df6.groupby('캐테고리')['구독자수'].agg(['count','sum'])
df7.columns = ['채널수','구독자수']
df7.sort_values(by='구독자수', ascending=False, inplace=True)
```

```python

plt.figure(figsize=(12,8))
sns.barplot(x='구독자수',y=df7.index, data=df7)
plt.title('카테고리별 구독자수')
plt.grid()
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.pie('구독자수', labels=df7.index, data=df7, autopct='%.1f%%')
plt.title('카테고리별 구독자 수')
plt.show()
```
