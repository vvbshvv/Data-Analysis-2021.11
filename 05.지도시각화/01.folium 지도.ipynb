```python
map=folium.Map(location=[35.23539781535266, 129.08051040639776], zoomstart=13)
folium.Marker(
    location=[35.23539781535266, 129.08051040639776], 
    tooltip='부산대학교', 
    popup=folium.Popup('부산광역시 금정구 장전2동 부산대학로63번길 2', max_width=200)
).add_to(map)

map
```

```python
#지도에 추가정보를 layer up해서 계속 쌓는 형식
map = folium.Map(location=[33.49982064711887, 126.53126472683314], zoom_start=13)
folium.Marker(
    location=[33.49982064711887, 126.53126472683314],
    tooltip='제주시청',
    popup=folium.Popup('제주시 광양9길 10', max_width=200),
).add_to(map)
folium.Marker(
    location=[33.451245414157576, 126.57218832683235], tooltip='제주테크노파크 에너지융합관',
    icon=folium.Icon(color='red', icon='info-sign')
).add_to(map)
folium.Circle(
    location=[33.51055660940559, 126.49136412683339], tooltip='제주공항',
    radius=500, color='red', fill=True
).add_to(map)
title = '<h3 align="center" style="font-size:20px">3개 지역 표시</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```
