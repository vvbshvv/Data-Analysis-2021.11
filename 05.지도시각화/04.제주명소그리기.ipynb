```python
import folium
import pandas as pd
```

```python
df=pd.read_csv('제주명소2.csv')
df
```

```python
map = folium.Map(location=[df.위도.mean(), df.경도.mean()], zoom_start=10)
for i in df.index:
    folium.Marker(
        location=[df.위도[i], df.경도[i]],
        tooltip=df.명소[i],
        popup=folium.Popup(df.도로명주소[i], max_width=200),
    ).add_to(map)

title = '<h3 align="center" style="font-size:20px">주요 명소 10개 지역 표시</h3>'
map.get_root().html.add_child(folium.Element(title))
map
```

```python
map.save('제주명소.html')
```

```python
import folium
from folium.features import CustomIcon

m = folium.Map(location=[33.48447359727584, 126.4897238826538], zoom_start=10)


icon_image1 = '222.jpg'

icon1 = CustomIcon(
    icon_image1,
    icon_size=(90, 80),
    icon_anchor=(10, 30),
    popup_anchor=(30, -30)
)

marker1 = folium.Marker(
    location=[33.48447359727584, 126.4897238826538],
    icon=icon1,
    popup=folium.Popup("차고지 지원 안되는 제주집", max_width='100')
)
m.add_child(marker1)

m
```
