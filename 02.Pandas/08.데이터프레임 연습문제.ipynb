붓꽃 종(species)별로 꽃잎길이(sepal_length), 꽃잎폭(sepal_width), 꽃받침길이(petal_length), 꽃받침폭(petal_width)의 평균, 표준편차 등 기초통계량(describe())을 구하시오.

```python
import seaborn as sns
iris = sns.load_dataset('iris')
iris.groupby('species')[['sepal_length','sepal_width','petal_length','petal_width']].describe()
```

3분위수(Q3)와 1분위수(Q1)의 차이보다 1.5배가 크거나 작은 데이터는 이상치이다. 즉,
Q1 - 1.5 * (Q3 - Q1) 보다 작은 데이터 Q3 + 1.5 * (Q3 - Q1) 보다 큰 데이터
이 이상치를 제거하고 위의 4가지 항목에 대해서 평균, 표준편차를 구하시오.

```python
iris.iloc[:,:]
```

```python
Q1=iris.quantile(.25)
Q3=iris.quantile(.75)
```

```python
q1=iris.groupby('species')['sepal_width'].quantile(.25)
q3=iris.groupby('species')['sepal_width'].quantile(.75)
iris['sepal_width']>q1-1.5*(q3-q1)
```

타이타닉호의 승객에 대해 나이와 성별에 의한 카테고리 열인 category1 열을 만드시오. category1 카테고리는 다음과 같이 정의됨
1) 20살이 넘으면 성별을 그대로 사용한다.
2) 20살 미만이면 성별에 관계없이 “child”라고 한다.

```python
titanic['category1']=titanic.apply(lambda x : 'child' if x.age<20 else x.sex, axis=1)
titanic
```
