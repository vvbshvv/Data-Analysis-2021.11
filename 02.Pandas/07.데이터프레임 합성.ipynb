```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
```

```python
df1 = pd.DataFrame({
'고객번호': [1001, 1002, 1003, 1004, 1005, 1006, 1007],
'이름': ['둘리', '도우너', '또치', '길동', '희동', '마이콜', '영희'] 
}, columns=['고객번호', '이름'])
df1
```

```python
df2 = pd.DataFrame({
'고객번호': [1001, 1001, 1005, 1006, 1008, 1001], '금액': [10000, 20000, 15000, 5000, 100000, 30000]
}, columns=['고객번호', '금액'])
df2
```

```python
# inner join 두개의 데이터프레임에서 같은 것만 골라 병합
df3 = pd.merge(df1, df2)
df3
```

```python
# how=outer(full outer join) 로 모든 데이터 병합 df1,df2 모두 가지고 있는거 병합
pd.merge(df1, df2, how='outer')
```

```python
# left outer join
pd.merge(df1, df2, how='left')# df1에만 있는거 만 플러스 병합해줌
```

```python
# right outer join
pd.merge(df1, df2, how='right')# df2에만 있는거 만 플러스 병합해줌
```

```python
df3=df1.set_index('고객번호')
df4 = df2.set_index('고객번호')
df3.join(df4)
```

- 서울시 인구 데이터와 cctv 데이터 합치기

```python
pop=pd.read_csv('서울시자치구별주민등록인구.txt', sep='\t', skiprows=2)
pop.head()
```

```python
pop = pop[['자치구','계','65세이상고령자']]
pop.head(3)
```

```python
pop.drop([0], inplace=True)
pop.head(3)
```

```python
pop.set_index('자치구', inplace=True)
pop.rename(columns={'계':'총인구', '65세이상고령자':'노인인구'}, inplace=True)
pop.head(3)
```

```python
pop=pop.apply(lambda x: x.str.replace(',','')).astype(int)
pop.head()
```

```python
pop['노인인구비율']=(pop['노인인구']/pop['총인구']*100).round(2)
```

```python
cctv=pd.read_csv('cctv.csv')
cctv.head(3)
```

```python
cctv.set_index('최근 증가율', inplace=True)
cctv.rename(columns={'최근 증가율':'노인인구비율'}, inplace=True)
cctv.head(3)
```

```python
df=pop.join(cctv)
df.head()
```

```python
#인구 천명당 cctv 댓수
df['천명당 cctv댓수'] = (df.cctv댓수/df.총인구*1000).round(2)
df.tail
```

```python
df.to_csv('서울시자치구별인구와cctv.csv')
```
